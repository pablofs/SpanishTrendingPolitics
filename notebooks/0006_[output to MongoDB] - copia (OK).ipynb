```python
def get_text_between( s, first, last, include_limits = False, first_ocurrences = True ):
    try:
        if first_ocurrences:
            if include_limits:
                start = s.index( first )
                end = s.rindex( last, start ) + len(last)
                return s[start:end]
            else:
                start = s.index( first ) + len( first )
                end = s.rindex( last, start ) 
                return s[start:end]
        else: 
            # look for last ocurrences of first and last
            if include_limits:
                start = s.rindex( first )
                end = s.index( last ) + len(last)
                return s[start:end]
            else:
                start = s.rindex( first ) + len( first )
                end = s.index( last ) 
                return s[start:end]            
    except ValueError:
        return ""
    
def remove_text_between( s, first, last, include_limits = False ):
    try:
        if include_limits:
            start = s.index( first ) + len( first )
            end = s.index( last, start ) 
            return s[:start] + s[end:] 
        else:
            start = s.index( first )
            end = s.index( last, start ) + len(last)
            return s[:start] + s[end:]
    except ValueError:
        return ""
    
def remove_spaces_and_newline(s):
    return re.sub(r'\s+', ' ', s.replace('\n',''))
```

```python
import re

def file_document_analyzer(text, date):

    prefix_file_name = date + "_doc"

    # Extraiem el text desde l'última aparició de 'PREGUNTAS:' i '<hr size="1" noshade="" />' (aquest últim element només apareix al final de cada doc)
    preguntas_html = get_text_between(text, 'PREGUNTAS:', '<hr size="1" noshade="" />', include_limits = False, first_ocurrences = False)

    # Cleaning the text of html tags:
    # Remove:
    # <p style="text-align:center"><a name="(Página7)"><b>Página7</b></a></p>
    preguntas_html = re.sub(r'<p.*?/p>', '', preguntas_html, flags=re.DOTALL)
    
    # Remove:
    # <center><a name="(Página8259)"><b>Página 8259</b></a></center>
    preguntas_html = re.sub(r'<center>.*?</center>', '', preguntas_html, flags=re.DOTALL)
    
    # No eliminem el tag <br /> si ens serveix per fer l' "split" en paràgrafs
    preguntas_html = re.sub(r'<br />', '', preguntas_html, flags=re.DOTALL)

    save_text_file(preguntas_html, 'log/' + prefix_file_name + '_preguntas_html.txt', "")
    #
    # paragraphs = preguntas_html.split("\n\n")
    # paragraphs = preguntas_html.split("<br />")
    paragraphs = preguntas_html.split("\n")
    text_to_analyze = ""
    i = 0
    for paragraph in paragraphs:
        text_to_analyze += "["+ str(i) + "]-----------------------------------------------------" + "\n"
        text_to_analyze += paragraph + "\n"
        i += 1
    # print text_to_analyze
    save_text_file(text_to_analyze, 'log/' + prefix_file_name + '_to_analyze.txt', "")

    #
    # paragraphs_analyzer(paragraphs)
    '''
    with Capturing() as output:
        result_dialogue_dict = paragraphs_analyzer(paragraphs)
    '''
    # result_dialogue_dict = my_paragraphs_analyzer(paragraphs)
    result_dialogue_dict = my_paragraphs_analyzer_v2(paragraphs)

    # Log output
    # print 'output:\n', output
    # save_list_text_file(output, 'log/' + prefix_file_name + '_analyzer_log.txt') 

    return result_dialogue_dict
```

```python
def get_group_color(groups_dict, group_code):
    if group_code:
        for key in groups_dict:
            if group_code == groups_dict[key]['code']:
                return groups_dict[key]['color']
    return 'cyan'

def mark_text_in_HTML(html_page, selected_text, start_index, group_color):
    text_start_index = html_page.find(selected_text, start_index)
    if text_start_index > 0:
        text_end_index = text_start_index + len(selected_text)
        html_page = html_page[:text_end_index] + tab_mark_end + html_page[text_end_index:]
        html_page = html_page[:text_start_index] + tab_mark_start.format(group_color) + html_page[text_start_index:]
    else:
        text_end_index = start_index

    return html_page, text_end_index


tab_mark_start = '<span style="background-color: {};">'
tab_mark_end = '</span>'

loaded_text_file = load_text_file('docs/20150311_doc.html',"")
# loaded_text_file = load_text_file('docs/19980429_doc.html',"")
loaded_groups_dict = load_dict_json('groups_dict.json')

text_index = loaded_text_file.rfind('PREGUNTAS:')

for num_dialogue in result_dialogue_dict:
    group_color = get_group_color(loaded_groups_dict, result_dialogue_dict[num_dialogue]['group'])
    old_text_index = text_index
    question_text = result_dialogue_dict[num_dialogue]['question'].strip()
    loaded_text_file, text_index = mark_text_in_HTML(loaded_text_file, question_text, text_index, group_color)
    if text_index == old_text_index:
        print "text_type:", 'question', "text_index:", text_index, ", num_dialogue:", num_dialogue
        print ">" + question_text + "<"
    for num_inter in result_dialogue_dict[num_dialogue]['intervention_dictonary']:
        paragraphs = result_dialogue_dict[num_dialogue]['intervention_dictonary'][num_inter]['text'].split(" <CONTINUED> ")
        
        id = result_dialogue_dict[num_dialogue]['intervention_dictonary'][num_inter]['who']
        if id != 9999:
            group = get_member_group_by_id(loaded_diputados_dict, id, 10)
            group_color = get_group_color(loaded_groups_dict, group)
        else:
            group_color = get_group_color(loaded_groups_dict, None)
            
        for paragraph in paragraphs:
            old_text_index = text_index
            paragraph_text = paragraph.strip()
            loaded_text_file, text_index = mark_text_in_HTML(loaded_text_file, paragraph_text, text_index, group_color)
            if text_index == old_text_index:
                print "text_type:", 'intervention', "text_index:", text_index, ", num_dialogue:", num_dialogue, "num_inter:", num_inter
                print ">" + paragraph_text + "<"

print "Saving 'docs/20150311_doc_marked.html'..."
save_text_file(loaded_text_file, 'docs/20150311_doc_marked.html',"")
# save_text_file(loaded_text_file, 'docs/19980429_doc_marked.html',"")
```
